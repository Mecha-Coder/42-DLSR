```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv("../dataset/dataset_train.csv")
houses = list(set(data["Hogwarts House"]))
courses = data.columns[6:]

print(houses)
print(courses, len(courses))
```

```python
fig, ax = plt.subplots(3, 5, figsize=(15, 10))

for i, c in enumerate(courses):
    
    plot = []
    for h in houses:
        plot.append(data[data["Hogwarts House"] == h][c])

    ax[int(i / 5), i % 5].hist(plot,
        bins=50, histtype='stepfilled', linewidth=1,
        color = ['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728'],
        edgecolor= ['#073354', '#a44807', '#0f490f', '#7a1414'],
        label = houses,
        alpha=0.6)
    
    ax[int(i / 5), i % 5].spines['top'].set_visible(False)
    ax[int(i / 5), i % 5].spines['right'].set_visible(False)
    ax[int(i / 5), i % 5].spines['left'].set_visible(False)
    ax[int(i / 5), i % 5].set_xlabel(c, fontsize=10)
    ax[int(i / 5), i % 5].set_xticks([])
    ax[int(i / 5), i % 5].set_yticks([])

fig.delaxes(ax[2,3])
fig.delaxes(ax[2,4])

handles, labels = ax[0, 0].get_legend_handles_labels()
fig.legend(handles, labels, loc='upper center', bbox_to_anchor=(0.75, 0.25), fontsize=12, ncol=2)
fig.suptitle('Hogwarts Courses Distribution by House', 
             fontsize=16, fontweight='bold', y=0.93)


ax[2,0].patch.set_facecolor('lightgray')
ax[0,0].patch.set_facecolor('lightgray')
```
