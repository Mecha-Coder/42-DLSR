```python
def predict(w, x, b):
    z = np.dot(w.T, x) + b
    return 1 / (1 + np.exp(-z))

def train(x, y, alpha=0.1, iter=500000, limit=0.000000000000001):
    
    m = x.shape[1]
    n = x.shape[0]

    w = np.zeros((n, 1))
    b = 0

    for i in range(iter):
        p = predict(w, x, b)
        loss = -(1/m) * np.sum(y*np.log(p) + (1-y)*np.log(1-p))

        if (i % 1000 == 0):
            print(f"Iteration: {i} | Loss: {loss}")
        
        step_w = alpha * (1/m) * np.dot(x, (p - y).T)
        step_b = alpha * (1/m) * np.sum(p - y)

        w = w - step_w
        b = b - step_b

        if (np.all(np.abs(step_w) <= limit) and np.abs(step_b) <= limit):
            print("Converge")
            break
        if (i == iter - 1):
            print("Didn't converge")

    
    return w, b
```
